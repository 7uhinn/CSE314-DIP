```python
A = np.zeros((1024,1024))
```

```python
c = 1
i,j = 0,0
while(i<1024):
    for j in range(1024):
        A[i][j] = 1
    i += 1  
    if(i==c*128):
        i += 128
        c += 2
```

```python
plt.imshow(A, cmap=plt.cm.bone)
plt.show()
```

### d. Create a grayscale image of size 256x1024. Intensity of image should vary sinusoidal.

```python
sin = np.zeros((256,1024))
for i in range(256):
    for j in range(1024):
        sin[i][j] = np.sin(j)
```

```python
plt.imshow(sin, cmap=plt.cm.bone)
plt.show()
```

### e. Create a white image of size 256x256, with black box of size 58x58 at centre.

```python
wibb = np.ones((256,256))
pad = (256-58)//2
for i in range(pad,(256-pad)):
    for j in range(pad,(256-pad)):
        wibb[i][j] = 0
```

```python
plt.imshow(wibb, cmap=plt.cm.bone)
plt.show()
```

### a. Image negative

```python
imggn = np.zeros((512,512))
for i in range(512):
    for j in range(512):
        imggn[i][j] = (255 - imgg[i][j])
```

```python
plt.imshow(imggn, cmap=plt.cm.bone)
plt.show()
```

### b. Log transformation and inverse log transform: s = c log (1+r), c is a const, r  0. s is pixel intensity of output image, r is the pixel intensity of input image. Study the effect of constant c on the quality of output image.

```python
imggl = np.zeros((512,512))
c = 1
for i in range(512):
    for j in range(512):
        imggl[i][j] = c*np.log(1+imgg[i][j])
```

```python
plt.imshow(imggl, cmap=plt.cm.bone)
plt.show()
```

```python
imggil = np.zeros((512,512))
c = 1
for i in range(512):
    for j in range(512):
        imggil[i][j] = (imgg[i][j])**(np.log(256) / (255)) - 1
```

```python
plt.imshow(imggil, cmap=plt.cm.bone)
plt.show()
```

### c. Power law transformation: Study the effect of different values of Gamma used in this transformation.

```python
imggp = np.zeros((512,512))
c = 1
for k in range(1,11):
    for i in range(512):
        for j in range(512):
            imggp[i][j] = c*(imgg[i][j])**k
    print("\nFor Gamma: ",k)
    plt.imshow(imggp, cmap=plt.cm.bone)
    plt.show()
```

### d. Contrast stretching

```python
a = 0
b = 255
c = np.amin(imgg)
d = np.amax(imgg)
```

```python
imggcs = np.zeros((512,512))
c = 1
for i in range(512):
    for j in range(512):
        imggcs[i][j] = (imgg[i][j]-c)*((b-a)/(d-c))+a
```

```python
plt.imshow(imggcs, cmap=plt.cm.bone)
plt.show()
```

### e. Gray level slicing

```python
min_range = 10
max_range = 60
```

```python
imgggls = np.zeros((512,512))
for i in range(512):
    for j in range(512):
        if imgg[i,j]>min_range and imgg[i,j]<max_range:
            imgggls[i,j] = 255
        else:
            imgggls[i,j] = 0
```

```python
plt.imshow(imgggls, cmap=plt.cm.bone)
plt.show()
```
