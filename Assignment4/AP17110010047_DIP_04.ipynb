#### Write a program to improve contrast of an image using histogram equalization. The prototype of the function is as below: 
#### histogram_equalisation(input_Image, no_of_bins);
#### The function should return the enhanced image.
#### Consider two low contrast input images. Study the nature of the output image quality in each case by varying the number of bins.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = plt.imread("lena_color.tiff")
plt.imshow(img, cmap=plt.cm.bone)
plt.show()
img.shape
```

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image, cmap=plt.cm.bone)
plt.show()
image.shape
```

```python
def histeq(image,bins):
    temp = 0
    rows,cols = image.shape
    oimg = np.zeros((rows,cols))
    ins = np.zeros(bins)
    cmd = np.zeros(bins)
    b = np.arange(256//bins,257,256//bins)

    for i in range(rows):
        for j in range(cols):
            ins[image[i][j]//(256//bins)] += 1
    
    pd = ins/(rows*cols)

    plt.bar(b,pd,width = 10)
    plt.xticks(b)
    plt.xlabel("Intensity Bins")
    plt.ylabel("Normalized Frequency")
    plt.title("Normalized Graph")
    plt.show()

    for i in range(bins):
        temp += pd[i]
        cmd[i] = np.round(temp*255)

    plt.bar(b,cmd,width = 50)
    plt.xticks(b)
    plt.xlabel("Intensity Bins")
    plt.ylabel("Normalized Frequency")
    plt.title("Cumultive Distribution")
    plt.show()

    for i in range(rows):
        for j in range(cols):
            oimg[i][j] = cmd[image[i][j]//(256//bins)]

    plt.imshow(oimg, cmap=plt.cm.bone)
    plt.show()
```

```python
arr=[2,4,8,16,32,64,128,256]
for i in arr:
    print("\nFor number of bins:",i,"\n")
    histeq(image,i)
```
