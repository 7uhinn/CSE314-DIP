```python
import numpy as np
```

```python
f = [7,13,20,8,5,2,1]
```

```python
for i in range(len(f)):
    print('T = ',i,'\n')
    print('Background: ')
    bw = []
    bms = []
    for j in range(0,i):
        bw.append(f[j])
        for p in range(f[j]):
            bms.append(j)
    wb = np.round(np.sum(bw)/np.sum(f),3)
    print('Wb = ',wb)
    mb = np.round(np.mean(bms),3)
    print('Mb = ',mb)
    sb2 = np.round(np.var(bms),3)
    print('Sb2 = ',sb2,'\n')
    
    print('Foreground: ')
    fw = []
    fms = []
    for k in range(i,len(f)):
        fw.append(f[k])
        for q in range(f[k]):
            fms.append(k)
    wf = np.round(np.sum(fw)/np.sum(f),3)
    print('Wf = ',wf)
    mf = np.round(np.mean(fms),3)
    print('Mf = ',mf)
    sf2 = np.round(np.var(fms),3)
    print('Sf2 = ',sf2,'\n')
    
    sw2 = np.round(wb*sb2 + wf*sf2,3)
    
    print('Sw2 = ',sw2,'\n\n')
```
