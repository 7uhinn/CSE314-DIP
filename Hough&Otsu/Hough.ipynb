```python
import numpy as np
import math
```

```python
p = [[4,2],[5,2],[6,2],[3,3],[7,3],[2,4],[8,4],[2,5],[6,5],[4,6],[2,7]]
```

```python
theta = []

for i in range(11):
    for j in range(i+1,11):
        if (p[i][0]-p[j][0] != 0):
            s = np.round((p[i][1]-p[j][1]) / (p[i][0]-p[j][0]),2)
            print(i,',',j,':',s)
            mc = np.round(math.atan(s),3)
            mcd = np.round(np.degrees(mc),0)
            if mcd not in theta:
                print(i,',',j,':',mcd)
                theta.append(mcd)
        else:
            mcd = 90.0
            if mcd not in theta:
                print(i,',',j,':',mcd)
                theta.append(mcd)
```

```python
trs = []
for i in range(len(theta)):
    rho = []
    for j in range(11):
        temp = p[j][0]*math.cos(np.deg2rad(theta[i]))+p[j][1]*math.sin(np.deg2rad(theta[i]))
        rho.append(np.round(temp,2))
    rho.sort(reverse=True)
    trs.append(rho)
```

```python
trs
```

```python
rhou = np.unique(trs)
```

```python
rhou
```

```python
f1 = []
f23 = []
tf = []
rf = []
for k in rhou:
    for i in range(len(theta)):
        count = 0
        for j in range(len(p)):
            if trs[i][j] == k:
                count += 1
        f1.append([theta[i],k,count])
        if count > 1:
            f23.append([theta[i],k,count])     
            tf.append(theta[i])
            rf.append(k)
```
